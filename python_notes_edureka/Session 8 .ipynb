```python
gr=df.groupby('Sex')
info=[]
for each,subset in gr:
    total=len(subset)
    survived_count=subset['Survived'].sum()
    info.append((each,total,survived_count))
info_df=pd.DataFrame(info,columns=['Group','Total','SurvivalCount'])
info_df['PercentageOfSurvial']=round(info_df['SurvivalCount']*100.0/info_df['Total'],2)

plt.barh(info_df['Group'],info_df['PercentageOfSurvial'])
plt.show()
info_df
```

```python
def Analysis(df,col_name):
    gr=df.groupby(col_name)
    info=[]
    for each,subset in gr:
        total=len(subset)
        survived_count=subset['Survived'].sum()
        info.append((str(each),total,survived_count))
    info_df=pd.DataFrame(info,columns=['Group','Total','SurvivalCount'])
    info_df['PercentageOfSurvial']=round(info_df['SurvivalCount']*100.0/info_df['Total'],2)

    plt.barh(info_df['Group'],info_df['PercentageOfSurvial'])
    plt.show()
    return info_df

```
